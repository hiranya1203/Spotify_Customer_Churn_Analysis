```python
# User Distribution by City
df["City"].value_counts().plot(kind="bar")

plt.title("User Distribution by City")
plt.xlabel("City")
plt.ylabel("Number of User")
plt.show()
```

```python
# The majority of users come from bangalore, mumbai, hyderabad, indicating higher adoption of streaming services.
```

```python
# Gender Distribution
df["Gender"].value_counts().plot(kind="pie", autopct = "%1.1f%%")
plt.title("Gender Distribution of Users")
plt.ylabel("")
plt.show()
```

```python
# From the gender distribution pie, the platform has a nearly balanced user base, with females representing about 51% of users and makes abotu 49%. this shows that spotify appeals to both the genders.
```

```python
# Device Type Usage
df["Device_Type"].value_counts().plot(kind="bar")

plt.title("Device Type Usage")
plt.xlabel("Device Type")
plt.ylabel("Number of Users")
plt.show()
```

```python
# In this analysis, users use more mobile devices than other, highlighting the importance of optimizing the mobile application experience.
```

```python
# Subscription Status
df["Subscription_Type"].value_counts().plot(kind="pie", autopct = "%1.1f%%")

plt.title("Subscription Status")
plt.ylabel("")
plt.show()
```

```python
# A large proportion of users remain active subscribers, while a smaller percentage have cancelled their subscriptions, indicating moderate churn
```

```python
# Plan Duration Distribution

df["Plan_Duration"].value_counts().plot(kind="bar")

plt.title("Plan Duration Distribution")
plt.xlabel("Plan Duration")
plt.ylabel("Number of Users")
plt.show()
```

```python
#Monthly plans are the most commonly chosen option, suggesting users prefer flexible subscription models.
```

```python
# Payment Method Distribution
df["Payment_Method"].value_counts().plot(kind="bar")

plt.title("Payment Method Distribution")
plt.xlabel("Payment Method")
plt.ylabel("Number of Users")
plt.show()
```

```python
# UPI appears to be the most preferred payment method, refelcting the increasing popularity of digital payments.
```

```python
# Daily Listening Minutes Distribution
plt.hist(df["Daily_Listening_Minutes"], bins = 20)
plt.title("Distribution of Daily Listening Minutes")
plt.xlabel("Daily Listening Minutes")
plt.ylabel("Number of Users")
plt.show()
```

```python
# The distribution shows that users have varied listening habits, with some users listening for short durations while other spending more time
```

```python
# Songa Played Per Day
plt.hist(df["Songs_Playes_Per_Day"], bins=20)

plt.title("Songs Played Per Day Distribution")
plt.xlabel("Songs Played Per Day")
plt.ylabel("Number of Users")
plt.show()
```

```python
# Majority of users play multiple songs daily, demostrating consistent content consumption on the platform
```

```python
# Churn vs Listening Minutes
sns.boxplot(x="Churn", y="Daily_Listening_Minutes", data=df)

plt.title("Churn vs Listening Minutes")
plt.show()
```

```python
# The listening minutes for churned and non churned users are quite similar.
```

```python
# Churn by Device Type
sns.countplot(x="Device_Type", hue="Churn", data =df)

plt.title("Churn by Device Type")
plt.show()
```

```python
# Most users access the platform through mobile device, and churn is also highest among mobile users due to the larger mobile user base.
```

```python
# Average Monthly Spend Distribution
plt.hist(df["Avg_Monthly_Spend"], bins=20)

plt.title("Average Monthly Spend Distribution")
plt.xlabel("Monthly Spend")
plt.ylabel("Number of Users")
plt.show()
```

```python
#The distribution shows that users spend varying amounts monthly on the platform, indicating different subscription preferences and spending behaviors among users.
```

```python
# Correlation Heatmap
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
